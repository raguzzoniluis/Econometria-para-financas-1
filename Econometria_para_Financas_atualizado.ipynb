### **Calculando retornos**

```python
BHP = yf.download(tickers = ('BHP'), start = '2003-1-1') [['Adj Close']]
XOM = yf.download(tickers = ('XOM'), start = '2003-1-1') [['Adj Close']]
PBR = yf.download(tickers = ('PBR'), start = '2003-1-1') [['Adj Close']]
VALE = yf.download(tickers = ('VALE'), start = '2003-1-1') [['Adj Close']]
```

```python
retorno_BHP = (BHP['Adj Close']/BHP['Adj Close'].shift(1)) -1
retorno_XOM = (XOM['Adj Close']/XOM['Adj Close'].shift(1)) -1
retorno_PBR = (PBR['Adj Close']/PBR['Adj Close'].shift(1)) -1
retorno_VALE = (VALE['Adj Close']/VALE['Adj Close'].shift(1)) -1
```

```python
retorno_medio_BHP = retorno_BHP.mean()
retorno_medio_BHP
```

```python
retorno_medio_PBR = retorno_PBR.mean()
retorno_medio_PBR
```

```python
retorno_medio_VALE = retorno_VALE.mean()
retorno_medio_VALE
```

```python
retorno_medio_XOM = retorno_XOM.mean()
retorno_medio_XOM
```

```python
retorno_anual_BHP = (retorno_medio_BHP * 252)
print(str(round(retorno_anual_BHP, 4)*100)+ '% a.a')
```

```python
retorno_anual_PBR = (retorno_medio_PBR * 252)
print(str(round(retorno_anual_PBR, 4)*100)+ '% a.a')
```

```python
retorno_anual_VALE = (retorno_medio_VALE * 252)
print(str(round(retorno_anual_VALE, 4)*100)+ '% a.a')
```

```python
retorno_anual_XOM = (retorno_medio_XOM * 252)
print(str(round(retorno_anual_XOM, 4)*100)+ '% a.a')
```

```python
retorno_log_BHP = np.log(BHP['Adj Close']/BHP['Adj Close'].shift(1))
retorno_log_XOM = np.log(XOM['Adj Close']/XOM['Adj Close'].shift(1))
retorno_log_PBR = np.log(PBR['Adj Close']/PBR['Adj Close'].shift(1))
retorno_log_VALE = np.log(VALE['Adj Close']/VALE['Adj Close'].shift(1))
```

```python
retorno_log_BHP.plot(figsize = (18,6))
```
